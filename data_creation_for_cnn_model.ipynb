```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import gc

import matplotlib
matplotlib.use('agg')


def window_average(arr):
    brr = []
    for i in range(4,len(arr)-4):
        window = np.array(arr[i-3:i+3])
        brr.append(window.mean())
    return np.array(brr)

for i in range(len(text_file)):
    cnt = 0
    r_peak = []
    label = []
    with open(f"/content/drive/My Drive/Colab Notebooks/mitbih_database/{text_file[i]}", 'r') as f:
        for line in f:
            cnt+=1
            line_as_string = str(line)
            p = line.split()
            if cnt>3:
                if p[2] in ["N","L","R","V","A"]:
                    r_peak.append(int(p[1]))
                    label.append(p[2])
    r_peak = r_peak[1:-1]
    label = label[1:-1]
    df = pd.read_csv(f"/content/drive/My Drive/Colab Notebooks/mitbih_database/{csv_file[i]}")
    voltage = window_average(df.iloc[:,2])
    segments = []
    for j in range(len(r_peak)):
        segments.append(voltage[r_peak[j]-150:r_peak[j]+150])
    print(csv_file[i],len(segments))
    cnt = 0
    for j in range(len(segments)):
        if i==20 and j<591:
          cnt+=1
          continue
        ecg = segments[j]
        fig = plt.figure(figsize=(4, 2.7), dpi=170)
        plt.plot(ecg)
        plt.axis('off')
        image_name = str(j)+"_"+csv_file[i]+"_"+label[j]+".png"
        folder = "/content/drive/My Drive/Colab Notebooks/cnn_ecg_image/"
        plt.savefig(folder+image_name)
        plt.clf()
        plt.close(fig)
        cnt+=1
        if cnt%100==0:
          gc.collect()
    gc.collect() 
```
